```python
import tensorflow as tf

from tensorflow.keras import Sequential, layers, callbacks, losses, metrics, optimizers

import matplotlib.pyplot as plt
```

```python
train, test = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((train[0].reshape(-1, 28, 28, 1)/ 255, tf.keras.utils.to_categorical(train[1])))
train_ds
```

```python
def show_im(xs):
    n = len(xs)
    for i, x in enumerate(xs):
        plt.subplot(1, n, i+1)
        plt.imshow(x)
        plt.axis("off")
    plt.show()    
```

```python
show_im(train_ds.take(5).map(lambda x, y: x))
```

```python

def get_model():
    model = Sequential()

    model.add(layers.Conv2D(filters=32, kernel_size = (3,3), activation='relu', input_shape=(28,28,1)))
    model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))

    model.add(layers.MaxPooling2D(pool_size=(2,2)))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(filters=64, kernel_size=(3,3),activation='relu'))
    model.add(layers.Conv2D(filters=64, kernel_size=(3,3),activation='relu'))    

    model.add(layers.MaxPooling2D(pool_size=(2,2)))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))

    model.add(layers.MaxPooling2D(pool_size=(2,2)))
    model.add(layers.Flatten())
    model.add(layers.BatchNormalization())
    model.add(layers.Dense(128, activation='relu'))

    model.add(layers.Dense(10, activation='linear'))
    loss = losses.CategoricalCrossentropy(from_logits=True)
    optim = optimizers.Adam(0.001)
    model.compile(
        loss=loss,
        optimizer = optim, 
        metrics = ["accuracy"]

    )
    return model 
```

```python
loss = losses.CategoricalCrossentropy(from_logits=True)
cbs = [
    callbacks.TensorBoard("./log/visv")
]
optim = optimizers.Adam(0.001)
```

```python
model.compile(
    loss=loss,
    optimizer = optim, 
    metrics = ["accuracy"]

)
model.summary()
```

```python
n = 1000
t = ((train[0].reshape(-1, 28, 28, 1)/ 255)[:n], tf.keras.utils.to_categorical(train[1])[:n])
```

```python
model_40 = get_model()
with_40 = model_40.fit(*t, epochs = 50, batch_size=40)
```

```python
def show_mets(history):
    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"])
    plt.xlim(-10, 60)
    plt.ylim(0, 1.10)
    plt.title("Accuracy")

    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"])
    plt.xlim(-10, 60)
    plt.ylim(-0.2, 2.5)
    plt.title("loss")
    
    plt.show()
```

```python
show_mets(with_40)
```

```python
model_1 = get_model()
with_1 = model_1.fit(*t, epochs = 50, batch_size=1)
show_mets(with_1)
```

```python
model_n = get_model()

with_n = model_n.fit(*t, epochs = 50, batch_size=n)

show_mets(with_n)
```
